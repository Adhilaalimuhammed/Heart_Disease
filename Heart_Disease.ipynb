# Applying MachineLearning Algorithms

```python
#feature selection
x = df.drop(['target'],axis=1)
y = df['target']
```

```python
#feature scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x = sc.fit_transform(x)
```

```python
#Splitting into traing and testing 
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(x,y,train_size=0.7,random_state=0) 
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lg_model = LogisticRegression()
lg_model.fit(X_train,y_train)
    
#predicted result
y_pred_lg = lg_model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm_lg = confusion_matrix(y_test,y_pred_lg) 

print(cm_lg)

#classification report
from sklearn.metrics import accuracy_score,classification_report
accuracy_score(y_test,y_pred_lg)

cp_lg = classification_report(y_test,y_pred_lg)
```

```python
print(cp_lg)
```

### DecisionTree

```python
from sklearn.tree import DecisionTreeClassifier
dt_model = DecisionTreeClassifier(criterion="gini")
dt_model.fit(X_train,y_train)

y_pred_dt = dt_model.predict(X_test)
```

```python
cm_dt =  confusion_matrix(y_test,y_pred_dt) 
cm_dt
```

```python
cp_dt = classification_report(y_test,y_pred_dt)
print(cp_dt)
```

### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rf_model = RandomForestClassifier(criterion = "gini",n_estimators=10,random_state=1)
rf_model.fit(X_train,y_train)

y_pred_rf = rf_model.predict(X_test)
```

```python
cm_rf = confusion_matrix(y_test,y_pred_rf)
cm_rf
```

```python
cp_rf = classification_report(y_test,y_pred_rf)
print(cp_rf)
```

### Neive bayes

```python
from sklearn.naive_bayes import GaussianNB
nbmodel = GaussianNB()
nbmodel.fit(X_train,y_train)

y_pred_nb = nbmodel.predict(X_test)
```

```python
cm_nb = confusion_matrix(y_test,y_pred_nb)
cm_nb
```

```python
cp_nb = classification_report(y_test,y_pred_nb)
print(cp_nb)
```

```python
print('lg: ',accuracy_score(y_test,y_pred_lg))
print('dt: ',accuracy_score(y_test,y_pred_dt))
print('rf: ',accuracy_score(y_test,y_pred_rf))
print('nb: ',accuracy_score(y_test,y_pred_nb))
```
